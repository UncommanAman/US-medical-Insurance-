# U.S. Medical Insurance Costs
In this project, I'm going to work on **insurance.csv** file which includes over 1000 data entries of US civilians regarding their different attributes i.e age, sex, bmi, number of children, etc. We will arrange the data in a more organised way and try to draw some insights.   

```python
import csv
```

Firstly, we have imported the required libraries, in this case we are working with .csv file so the only required library which we need is csv library.

Looking at the **insurance.csv** file, we can broadly divide the data into seven fields. These fields are :
* Age of the Person
* Sex of the person
* BMI of the person
* Number of Children
* Smoker Status 
* Region the person live in
* Yearly Medical insurance cost

We will now make seven different emplty list to accumulate the data from the each column of **insurance.csv** file.

```python
age = []
sex = []
bmi = []
num_of_children = []
smoker_status = []
region = []
insurance_cost = []
```

Now that we have created seven empty lists for our data, we can open the ** insurance.csv ** file with the help of our csv library that we earlier imported. To ease the prcoess, we can make a function to append the data into each list rather than doing the same task seprately for sevent times.  

```python
def load_data_file(csv_file, lst, columnn_name):
    with open(csv_file) as csv_obj:
        csv_dict = csv.DictReader(csv_obj)
        for row in csv_dict :
            lst.append(row[columnn_name])
    return lst
```

Now as we have preprared the fucntion to read each column from the insurance.csv file, we can call the function seven times by passing all the column names as arguments. We then have stored the all the data into new list which we made eariler.  

```python
age = load_data_file('insurance.csv',age , 'age')
sex = load_data_file('insurance.csv',sex , 'sex')
bmi = load_data_file('insurance.csv',bmi , 'bmi')
num_of_children = load_data_file('insurance.csv',num_of_children , 'children')
smoker_status = load_data_file('insurance.csv',smoker_status , 'smoker')
region = load_data_file('insurance.csv',region , 'region')
insurance_cost = load_data_file('insurance.csv',insurance_cost , 'charges')
```

We can see the data in each of the list by passing each list into print() statement.

```python
print(age)
print(sex)
print(bmi)
print(num_of_children)
print(smoker_status)
print(region)
print(insurance_cost)
```

Now it's time to decide what we have to do with the data we retrived from the file. For this project we are going to find some of the insights from the data. 
* Average age of the paitents
* Difference between average insurance cost for the people who smoke and for the people who don't 
* Analyzing where the most people are from
* Creating a dictinoary for all the paitence's details

We will be creating different functions to carry out our tasks which are mentioned above. Let's first start with finding the average age of our paitents. 

```python
def average_age(lst):
    total_age = 0
    for every_age in lst:
        total_age += int(every_age)
        average_age = total_age / len(lst)
    return average_age 
```

```python
average_age = average_age(age)
print(round(average_age))
```

Now that we have sucesfully found the average age of all our paitents, we can move on to see how being a smoker affects the insurance cost of the person. For this we will find the average insurance cost of every person who smokes and compare it with the average insurance cost of the people who don't smoke. Depending upon the difference, we can find how smoking affects the insurance cost of a person. 


Since every data stored in each list correspond to the data in the other list, we can simply deduce the affect with the help of our insurance_cost list and smoker_status list. We will start by building a function name *smoking_diff()*

```python
def smoking_diff(lst1, lst2 ):
    avg_smoker_cost = 0
    avg_non_smoker_cost = 0
    i = 0 
    j = 0
    for value1, value2 in zip(lst1, lst2):
        if value1 == 'yes':
            avg_smoker_cost += float(value2)
            i += 1
        elif value1 == 'no':
            avg_non_smoker_cost += float(value2)
            j += 1
    avg_smoker_cost /= i
    avg_non_smoker_cost /= j
    print('People who smoke have an average insurance cost of {avg1} and people who do not smoke have an average insurance cost of {avg2}.'.format(avg1 = round(avg_smoker_cost, 2), avg2 = round(avg_non_smoker_cost, 2)))
    print('The difference between the insurance cost of people who smoke and people who do not is :',abs(round(avg_smoker_cost - avg_non_smoker_cost, 2)))
```

```python
smoking_diff(smoker_status, insurance_cost)
```

Now we are going to find the the region that most number of the paitents belong to. For this we will create a seprate function which will return the region with maximum number of paitents. 

```python
def max_resi(lst):
    max = 0
    res = lst[0]
    for x in lst:
        freq = lst.count(x)
        if freq > max:
            max = freq
            res = x
    print('Most number of people belong to:', res)
```

```python
max_resi(region)
```

As we observed from our anlaysis, maximum number of people in our data set belong to southeast. 

Now we are going to make a dictionary to store all the data we have retrieved so far from our dataset *insurance.csv*. 

```python
def create_dict(list_age, list_sex, list_bmi, list_children, list_smoker, list_region, list_cost):
    new_dict2 = {}
    new_dict3 = {}
    for x in range(1,len(list_age)):
        new_dict2['Age'] = list_age[x]
        new_dict2['Sex:'] = list_sex[x]
        new_dict2['BMI'] = list_bmi[x]
        new_dict2['Number of Children'] = list_children[x]
        new_dict2['Smoker Status'] = list_smoker[x]
        new_dict2['Region:'] = list_region[x]
        new_dict2['Insurance Cost:'] = list_cost[x]
        new_dict3[x] = new_dict2
    return new_dict3
```

```python
paitent_dict = create_dict(age, sex, bmi, num_of_children, smoker_status, region, insurance_cost )
print(paitent_dict)
```
